```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tiempo = np.loadtxt("salida1.txt", usecols = 0)
theta = np.loadtxt("salida1.txt", usecols = 1)
phi = np.loadtxt("salida1.txt", usecols = 2)
```

```python
tiempo2 = np.loadtxt("salida2.txt", usecols = 0)
theta2 = np.loadtxt("salida2.txt", usecols = 1)
phi2 = np.loadtxt("salida2.txt", usecols = 2)
```

```python
plt_1 = plt.figure(figsize=(6, 5))
plt.plot(theta, phi)
plt.scatter(theta[0], phi[0], label = "Inicio trayectoria 1")
plt.scatter(theta[749], phi[749], label = "Final trayectoria 1")
plt.plot(theta2, phi2)
plt.scatter(theta2[0], phi2[0], label = "Inicio trayectoria 2")
plt.scatter(theta2[749], phi2[749], label = "Final trayectoria 2")
plt.legend()
plt.title("Trayectorias en el espacio fase")
plt.xlabel(r"$\theta$")
plt.ylabel(r"$\phi$")
```

```python
plt_2 = plt.figure(figsize=(6, 5))
plt.plot(tiempo, phi)
# Esto funciona porque son datos con misma dt y misma cantidad de pasos, aguas
plt.plot(tiempo2, phi2)
plt.title("Trayectoria del angulo 2 en el tiempo")
plt.xlabel(r"$t$")
plt.ylabel(r"$\phi$")
plt.savefig('angulo2.png')
```

```python
plt_2 = plt.figure(figsize=(6, 5))
plt.plot(tiempo, theta)
plt.plot(tiempo2, theta2)
plt.title("Trayectoria del angulo 1 en el tiempo")
plt.xlabel(r"$t$")
plt.ylabel(r"$\theta$")
plt.savefig('angulo1.png')
```
